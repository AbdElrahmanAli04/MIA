```python
import os 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns 
#import seaborn as sns


```

```python
mydf = pd.read_csv("D:\Programming\TASK4.1\Weather Dataset.csv")
mydf = mydf.rename(columns={ "Loud Cover" : "Cloud Cover" })
mydf.columns 
```

```python
mydf.describe() 
#handle missng vales ?
```

```python
#Handeled false variables 
mydf = mydf[(mydf["Temperature (C)"] > -273) & (mydf["Humidity"] >= 0) & (mydf["Visibility (km)"] >= 0) & (mydf["Cloud Cover"] >= 0)]
mydf.describe()
```

```python
#Removing Duplicates
mydf = mydf.drop_duplicates(subset=["Formatted Date"])
```

```python
#Now the data is cleand
mydf
```

```python
#Ddrop not necessary data
mydf = mydf.drop(columns="Cloud Cover")
mydf
```

```python
mydf["Formatted Date"] = mydf["Formatted Date"].apply(lambda x: str(x))  # Convert to string
mydf["Formatted Date"] = mydf["Formatted Date"].apply(lambda x: x[0:4] + '-' + x[5:7] + '-' + x[8:10])  # Format as XXXX-XX-XX
mydf
```

```python
#Time-Temp Plot
x = mydf["Formatted Date"].values
y = mydf["Temperature (C)"].values
plt.plot(x,y)
plt.xlabel("Time")
plt.ylabel("Temperature (C)")
plt.show()
```

```python
#Histogram Temp dist
Temp = mydf["Temperature (C)"].values

plt.hist(Temp)
plt.show()
```

```python
#Scatter of Temp Vs Humidty
plt.figure(figsize=(15,10))
Temp = mydf["Temperature (C)"].values
Humidty = mydf["Humidity"].values
plt.scatter(Temp,Humidty,s=10,c="green")
plt.show

```

```python
#correlation heatmap
heatdf = mydf.drop(columns=["Formatted Date", "Daily Summary"  , "Precip Type" , "Summary" ])
heatdf
corr_matrix = heatdf.corr()
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', center=0)
plt.title('Correlation Heatmap')
plt.show()
```
